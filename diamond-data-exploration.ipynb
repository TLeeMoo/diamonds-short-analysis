```r
library(tidyverse)
```

```r
###Preview & Summarize Data
head(diamonds)
str(diamonds)
glimpse(diamonds)
```

```r
###List Column Names
colnames(diamonds)
```

```r
### Cleaning stage: Change column names x=length mm y=width mm
rename(diamonds, widthmm=y)
```

```r
###Use summarize functions to get stats on avg carat and price
summarize(diamonds, mean_carat=mean(carat))
```

```r
summarize(diamonds,mean_price=mean(price))
```

```r
###From results above, the average weight for the diamond is about 0.798 carats and the average price is $3,932.80USD
###Visualize data: 1.Scatterplot
ggplot(data = diamonds, aes(x = carat, y = price)) +
  geom_point()
```

```r
ggplot(data = diamonds, aes(x = carat, y = price, color = cut)) +
  geom_point()
```

```r
###Simplify Graph by creating plot for each cut using facet_wrap function
ggplot(data = diamonds, aes(x = carat, y = price, color = cut)) +
  geom_point() +
  facet_wrap(~cut)
```
